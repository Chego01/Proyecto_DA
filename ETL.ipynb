Importamos las librerias necesarias para el proceso ETL

```python
import pandas as pd 
import os
```

Realizamos el proceso de extracción de archivos para convertirlos en dataframes

```python
# Ruta al archivo XLSX
hom_archivo = 'Datasets\homicidios.xlsx'
les_archivo = 'Datasets\lesiones.xlsx'
sini_archivo = 'Datasets\siniestros_2023.xlsx'
mujeres_archivo = 'Datasets\mujeres_heridas.xlsx'
# Lee el archivo XLSX y carga su contenido en un DataFrame
df_hom = pd.read_excel(hom_archivo)
df_les = pd.read_excel(les_archivo)
df_siniestros = pd.read_excel(sini_archivo)
df_mujeres = pd.read_excel(mujeres_archivo)
```

Leemos el archivos de homicidios.xlsx

```python
df_hom.info()
```

Se procede a manipular columnas para su eliminación,renombramiento y reset de indice

```python
#Columnas a eliminar
columns_to_drop = ['LUGAR_DEL_HECHO','Calle','XY (CABA)','PARTICIPANTES','Altura']
df_hom = df_hom.drop(columns=columns_to_drop)
#Columnas para renombrar
df_hom = df_hom.rename(columns={'ID':'Id','N_VICTIMAS':'Victimas','AAAA':'Año','MM':'Mes','DD':'Dia','TIPO_DE_CALLE':'Tipo_calle','HH':'Franja_Horaria','Dirección Normalizada':'Direccion_Normalizada','pos x': 'X', 'pos y': 'Y', 'VICTIMA': 'Victima','ACUSADO':'Acusado'})
# Obtén el valor mínimo actual de la columna 'ID'
min_id = int(df_hom['Id'].str.split('-').str[-1].min())
# Resta el valor mínimo de la columna 'ID' de todos los valores para comenzar desde 0
df_hom['Id'] = (df_hom['Id'].str.split('-').str[-1].astype(int) - min_id).astype(str)
```

Presentación final del archivo de homicidos.xlsx

```python
df_hom.head(5)
```

```python
df_hom_victimas= pd.read_excel(hom_archivo,sheet_name="VICTIMAS")
```

```python
#Columnas a eliminar
columns_to_drop = ['FECHA_FALLECIMIENTO','MM','DD','FECHA','AAAA','VICTIMA']
df_hom_victimas = df_hom_victimas.drop(columns=columns_to_drop)
#Columnas para renombrar
df_hom_victimas = df_hom_victimas.rename(columns={'ID_hecho':'Id','ROL':'Rol', 'SEXO': 'Sexo','EDAD':'Edad'})
# Obtén el valor mínimo actual de la columna 'ID'
min_id = int(df_hom_victimas['Id'].str.split('-').str[-1].min())
# Resta el valor mínimo de la columna 'ID' de todos los valores para comenzar desde 0
df_hom_victimas['Id'] = (df_hom_victimas['Id'].str.split('-').str[-1].astype(int) - min_id).astype(str)
```

```python
df_hom_victimas
```

```python
data_homicidios = df_hom.merge(df_hom_victimas,on='Id',how='left')
```

```python
data_homicidios
```

```python
# Guardar el DataFrame como archivo CSV en la ruta especificada
data_homicidios.to_csv("EDA/Datasets/homicidos.csv", index=False)
```

Leemos el archivo lesiones.xlsx

```python
df_les.info()
```

Procedemos a realizar un reset del indice al igual que eliminar,renombrar y cambiar el orden de las columnas 

```python
#Reset de indice 
df_les.reset_index(inplace=True)
#Columnas a eliminar
columns_to_drop = ['id','mm','dd','fecha','hora','franja_hora','comuna','otra_direccion','calle','altura','cruce','geocodificacion_CABA','participantes','moto','auto','transporte_publico','camion','ciclista','gravedad']
df_les = df_les.drop(columns=columns_to_drop)
#Columnas a renombrar
df_les = df_les.rename(columns={'index':'Id','n_victimas':'Victimas','aaaa':'Año','direccion_normalizada':'Direccion_Normalizada','tipo_calle':'Tipo_calle','longitud':'X','latutid':'Y','victima':'Victima','acusado':'Acusado'})
order_columns = ['Id','Victimas','Año','Tipo_calle','Direccion_Normalizada','X','Y','Victima','Acusado']
df_les = df_les[order_columns]
```

Presentación final del archivo lesiones.xlsx

```python
df_les.head(5)
```

```python
df_les_victimas = pd.read_excel(les_archivo,sheet_name="VICTIMAS")
```

```python
#Reset de indice 
df_les_victimas.reset_index(inplace=True)
#Columnas a eliminar
columns_to_drop = ['ID hecho','MM','DD','FECHA ','VEHICULO_VICTIMA','AAA']
df_les_victimas = df_les_victimas.drop(columns=columns_to_drop)
#Columnas a renombrar
df_les_victimas = df_les_victimas.rename(columns={'index':'Id','SEXO':'Sexo','EDAD_VICTIMA':'Edad','GRAVEDAD':'Gravedad'})
order_columns = ['Id','Victimas','Año','Tipo_calle','Direccion_Normalizada','X','Y','Victima','Acusado']
df_les_victimas = df_les_victimas[order_columns]
```

```python
data_lesiones = df_les.merge(df_les_victimas,on='Id',how='left')
```

```python
data_lesiones
```

```python
# Guarda el DataFrame en el archivo CSV
data_lesiones.to_csv('EDA\Datasets\Lesiones.csv',index=False)
```

Lectura del archivo siniestros.xlsx

```python
df_siniestros.info()
```

Procedemos a manipular las columnas para poder extraer los datos necesarios, renombre, eliminación y reordenamiento de columnas

```python
#Cambio de tipo de dato de la columna FECHA
df_siniestros['FECHA'] = df_siniestros['FECHA'].astype(str)
df_siniestros['Año'] = df_siniestros['FECHA'].str.slice(0, 4)
df_siniestros['FECHA'].str.extract(r'(\d{4})') 
#Generación de nueva columna
df_siniestros['Victimas']= df_siniestros['MOTO']+df_siniestros['LVIANO']+df_siniestros['BUS']+df_siniestros['CAMION']
#Columnas a eliminar 
columns_to_drop = ['FECHA','HORA','BANDA y/o RAMAL','PK','CONDICIONES METEOROLÓGICAS','SUPERFICIE DE LA VIA','TIPO DE SINIESTRO','LESIONADOS','FALLECIDOS','MOTO','LVIANO','BUS','CAMION']
df_siniestros = df_siniestros.drop(columns=columns_to_drop)
#Creación de columnas 
df_siniestros['Acusado'] = 'SD'
df_siniestros['Tipo_calle'] = 'SD'
df_siniestros['Victima'] = 'SD'
#Recombramiento de columnas
df_siniestros = df_siniestros.rename(columns={'AUTOPISTA':'Direccion_Normalizada'})
#Cambio de orden de columnas
new_order_columns = ['Victimas','Año','Tipo_calle','Direccion_Normalizada','Victima','Acusado']
df_siniestros=df_siniestros[new_order_columns]
```

```python
#Reset de indice
df_siniestros.reset_index(inplace=True)
```

```python
#Renombre de columnas del dataframe
df_siniestros = df_siniestros.rename(columns={'index':'Id'})
```

```python
# Guarda el DataFrame en el archivo CSV
df_siniestros.to_csv('EDA\Datasets\siniestros.csv',index=False)
```
